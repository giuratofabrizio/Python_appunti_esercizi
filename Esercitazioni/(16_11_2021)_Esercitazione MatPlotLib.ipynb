```python
import pandas as pd
```

1

```python
d = {
    "last_name": ['Boyer','Copeland','Daniel','Serrano','Terrel','Vargas'],
    "orders_number": [185, 50, 196, 58, 24, 28]
}

df = pd.DataFrame(d)
df
```

2

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes()
ax.bar(df['last_name'], df['orders_number'])
```

3

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df['last_name'], df['orders_number'], color='r')

ax.set_xlabel('Commessi')
ax.set_ylabel('Ordini')
fig.suptitle('Numero di ordini per ogni commesso')
```

4

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df.sort_values('orders_number', ascending= False)['last_name'],df['orders_number'].sort_values(ascending= False), color='b')  #VALORI IN ORDINE NUMERICO
```

5

```python
fig = plt.figure()
ax = plt.axes()
ax.barh(df['last_name'], df['orders_number'], color='r')  #barh per horizontal bars

ax.set_xlabel('Commessi')
ax.set_ylabel('Ordini')
fig.suptitle('Numero di ordini per ogni commesso')
```

6

```python
tot = 185+50+196+58+24+28

d2 = {
    'last_name': ['Copeland','Serrano','Boyer','Daniel','Vargas','Terrell'],
    'order_perc': [round((185*100)/tot,2), round((50*100)/tot,2), round((196*100)/tot,2), round((58*100)/tot,2), round((24*100)/tot,2), round((28*100)/tot,2)]
}

df2=pd.DataFrame(d2)
df2
```

 7/8

```python
slices = list(df2['last_name'])
activities = list(df2['order_perc'])
plt.rcParams.update({'font.size':25})
fig = plt.figure(figsize=[7,7])
ax = plt.axes()

#CREAZIONE GRAFICO A TORTA
ax.pie(activities,labels=slices,autopct='%.2f')
plt.show()
```
