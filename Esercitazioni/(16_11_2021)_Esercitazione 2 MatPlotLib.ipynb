```python
import pandas as pd
import matplotlib.pyplot as plt

d2 = {
    'last_name': ['Boyer','Copeland','Daniel','Serrano','Terrel','Vargas'],
    'quantity': [766,213,812,269,100,114],
    'total_amount': [643049.37, 203549.39, 744259.20, 253183.68, 93851.95, 123156.90]
}

df =pd.DataFrame(d2)
df
```

B

```python
x1 = list(df['last_name'])
y1 = list(df['quantity'])
y2 = list(df['total_amount'])


fig = plt.figure(figsize=(18,8))
ax1 = plt.subplot(1,2,1)                 #Grafici affiancati posizione 1
ax2 = plt.subplot(1,2,2)                 #Grafici affiancati posizione 2

ax1.bar(x1,y1)
ax2.bar(x1,y2)
```
