```python
gg = int(input('inserire giorno: '))
mm = int(input('inserire mese: '))
aaaa = int(input('inserire anno: '))

if gg > 31:
  print("giorno selezionato non disponibile")
elif mm > 12:
  print("mese selezionato non disponibile")
elif aaaa > 2021:
  print('anno selezionato non disponibile')

print(gg, '/', mm, '/', aaaa)
```

```python
listaParole = [elemento for elemento in input('inserire parole: ').split()]

letteraUser = input('inserire lettera: ')

print(listaParole)
for parola in listaParole:
  if parola[0] == letteraUser:
     listaParole.remove(parola)

print(listaParole)
```
