```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/somministrazioni-vaccini-latest.csv')
df
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/platea.csv')
df1
```

```python
df1.groupby('area')['totale_popolazione'].value_counts().unstack(0).plot.bar()

#Creazione di doppie assi per ogni indice come fatto nell'esercizio del 24/11
```

```python
perc = df1.groupby('area')['totale_popolazione'].value_counts(normalize=True)

# perc = valore percentuale della popolazione totale per ogni area(groupby)


fig = plt.figure()
ax = plt.axes()

ax.bar(df1.index, perc.values, color='r')

#Creazione del grafico con indici di DF1 e valori in percentuale presi da perc
```
