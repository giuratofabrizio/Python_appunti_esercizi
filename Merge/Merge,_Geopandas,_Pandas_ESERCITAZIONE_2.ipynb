```python
#  converto la colonna ID_NIL da int64 a object(str)
quartieri_milano["ID_NIL"] = quartieri_milano["ID_NIL"].astype(str)
quartieri_milano.dtypes
```

```python
area_sosta.dtypes
```

```python
#  unione tra quartieri_milano e area_sosta in base al: left_on="ID_NIL" e right_on="num_civico"
quartieri_milano_area_sosta = pd.merge(quartieri_milano, area_sosta, how="outer", left_on="ID_NIL", right_on="num_civico")
quartieri_milano_area_sosta
```

```python
# input
nil_inserito = input("inserire nome del quartiere: ")
# localizzazione del quartiere
dati_dei_quartieri_inseriti = quartieri_milano_area_sosta[quartieri_milano_area_sosta.NIL == nil_inserito]
dati_dei_quartieri_inseriti
```

```python
info_quartiere = quartieri_milano[quartieri_milano.NIL == nil_inserito]
info_quartiere
```

```python
info_quartiere_ID_NIL = info_quartiere[["ID_NIL"]]
info_quartiere_ID_NIL
```
